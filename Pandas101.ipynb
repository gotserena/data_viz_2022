#Quiz (ให้ใช้เฉพาะที่อาจารย์สอนไปแล้วในวิชานี้)

1.   คำนวณ อายุเฉลี่ย ของ**ผู้ชาย** ในข้อมูลทั้งหมด
2.   คำนวณ อายุเฉลี่ย ของ**ผู้ชาย** ในจังหวัดขอนแก่น

####ข้อ1  คำนวณ อายุเฉลี่ย ของ**ผู้ชาย** ในข้อมูลทั้งหมด

````python
this_data = data_covid[['sex','age','province_of_onset']] #กำหนดตัวแปร และโชว์ข้อมูล sex age province_of_onset
````

````python
this_data #โชว์ตาราง
````

````python
men = this_data[this_data['sex']=='ชาย'] #กำหนดเพศเอาแต่ผู้ชาย
men['age'] #โชว์
````

````python
sum = 0 #นับเริ่มจาก 0
N = 0 #กำหนด Nเริ่ม คือ 0
for a in men['age']: #วนลูป ให้ a คืออายุชาย
    
    if a > 0: 
        sum += a # sum = sum + a เพื่อ ให้ค่าใหม่มาบวกกับsumเก่าเรื่อย ๆ
        N += 1 #จำนวนข้อมูลจะเพิ่มเรื่อย ๆ ตามจำนวนที่มีเหลือ

print(f'อายุเฉลี่ย ของ ผู้ป่วยชาย {sum/N}')
````

###ข้อ2 คำนวณ อายุเฉลี่ย ของ**ผู้ชาย** ในจังหวัดขอนแก่น

````python
men_Khonkaen = men[men['province_of_onset']=='ขอนแก่น'] #ให้ province_of_onset เอาแต่ ขอนแก่น
sum = 0 #นับเริ่มจาก 0
N = 0 #กำหนด Nเริ่ม คือ 0
for a in men_Khonkaen['age']:#วนลูป ให้ a คืออายุชาย โดยจะเหลือแค่ province_of_onset ที่เป็นขอนแก่น
    
    if a > 0:
        sum += a # sum = sum + a เพื่อ ให้ค่าใหม่มาบวกกับsumเก่าเรื่อย ๆ
        N += 1 #จำนวนข้อมูลจะเพิ่มเรื่อย ๆ ตามจำนวนที่มีเหลือ

print(f'อายุเฉลี่ย ของ ผู้ป่วยชาย ในจังหวัดขอนแก่น {sum/N}')
````

### การวนลูป record ในตาราง  (```.iterrows```)

````python
this_data = data_covid[['sex','age','province_of_onset']]
````

````python
for each_row in this_data.iterrows():
    # print(each_row)
    # print(each_row[1])
    if (each_row[1]['age'] == 20) and (each_row[1]['province_of_onset'] == 'ขอนแก่น'):
        print(each_row)
````

### การวนลูป แบบมองตารางแพนด้าส์(pandas dataframe) เป็น numpy array  (```.iloc```)

````python
for each_row in range(this_data.shape[0]):
    if (this_data.iloc[each_row,1] == 20) and (this_data.iloc[each_row,2] == 'ขอนแก่น'):  #each_row คือ ตัวเลข
        print(each_row)
        print(this_data.iloc[each_row,:])   
````

## Quiz ตัดตารางออกมาเฉพาะปี 2021 ```announce_date``` ในปี 2021 

**Hint**
*   วนลูปหา index ของปี 2021
*   ตัดตารางมาเฉพาะ ปี 2021



````python
TF=list()
for each_row in data_covid.iterrows():
    if each_row[1]['announce_date'].split('/')[2] == '2021':
        TF.append(True)
    else:
        TF.append(False)

data_covid[TF].head()
````
